```python
def extract_movie_stats(df):
    stats = df[['movie', 'rating']].groupby(by='movie').agg({'rating': ['count', 'mean']}).reset_index()
    stats = stats.droplevel(0, axis=1)
    stats.rename(columns={'count': 'review_count', 'mean': 'rating_mean', '': 'movie'}, inplace=True)
    stats['review_share'] = stats['review_count'] / len(df)
    stats['rating_mean'] /= stats['rating_mean'].max()
    stats.sort_values(by='review_share', inplace=True)
    return stats
```

```python
anon_movie_stats = extract_movie_stats(anon_df)
imdb_movie_stats = extract_movie_stats(imdb_df)
```
